```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = open("stats_fix_P.txt", "r")
data = f.read()
f.close()
T = list(map(float, data.split()))

f = open("stats_fix_P1.txt", "r")
data = f.read()
f.close()
T1 = list(map(float, data.split()))

N = [i*200 for i in range(0, len(T))]
S = [T1[i]/T[i] for i in range(len(T))]
E = [S[i]/4 for i in range(len(S))]

plt.plot(N, T)
plt.title('Зависимость времени от числа частиц при P = 4')
plt.show()



```

```python
plt.plot(N, S)
plt.title('Зависимость ускорения от числа частиц при P = 4')
plt.show()
```

```python
plt.plot(N, E)
plt.title('Зависимость эффективности от числа частиц при P = 4')
plt.show()
```

```python
f = open("stats_fix_N.txt", "r")
data = f.read()
f.close()
T = list(map(float, data.split()))

N = [i+1 for i in range(0, len(T))]
S = [T[0]/T[i] for i in range(len(T))]
E = [S[i]/N[i] for i in range(len(S))]

plt.plot(N, T)
plt.title('Зависимость времени от числа потоков при N = 50000')
plt.show()

plt.plot(N, S)
plt.title('Зависимость ускорения от числа потоков при N = 50000')
plt.show()

plt.plot(N, E)
plt.title('Зависимость эффективности от числа потоков при N = 50000')
plt.show()
```
