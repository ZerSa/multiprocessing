```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = open("time_P.txt", "r")
data = f.read()
f.close()
T = list(map(float, data.split()))
Q = [0.137254 for i in range(5)]
P = [2**i for i in range(5)]
T_omp = [0.167545, 0.0833, 0.085076, 0.112135, 0.08444] 


plt.plot(P, T, label='pthread_sort')
plt.plot(P, T_omp, label='omp_sort')
plt.plot(P, Q, label='qsort')
plt.legend()
plt.title('Зависимость времени от числа потоков')
plt.show()
```

Все графики и данные для n = 1000000

```python
S = [T[0]/T[i] for i in range(5)]
E = [S[i]/P[i] for i in range(5)]

plt.plot(P, S)
plt.title('Зависимость ускорения от числа потоков')
plt.show()

plt.plot(P, E)
plt.title('Зависимость эффективности от числа потоков')
plt.show()
```
