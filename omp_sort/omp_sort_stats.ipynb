```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = open("time_m.txt", "r")
data = f.read()
f.close()
T = list(map(float, data.split()))
M = [0.01*i for i in range(1, 101)]

plt.plot(M, T)
plt.title('Зависимость времени от размера чанков(в процентах от числа элементов)')
plt.show()
```

Возьмем размер чанка = 30% от числа элементов

```python
f = open("time_P.txt", "r")
data = f.read()
f.close()
T1 = list(map(float, data.split()))
Q = [T1[-1] for i in range(5)]
P = [2**i for i in range(5)]
print(T1)
T1 = T1[:5]
plt.plot(P, T1)
plt.plot(P, Q)
plt.title('Зависимость времени от числа потоков')
plt.show()
```

График построен при n = 1000000

```python
S = [T1[0]/T1[i] for i in range(5)]
E = [S[i]/P[i] for i in range(5)]

plt.plot(P, S)
plt.title('Зависимость ускорения от числа потоков')
plt.show()

plt.plot(P, E)
plt.title('Зависимость эффективности от числа потоков')
plt.show()
```
